```python
year_pattern = r'\b(19|20)\d{2}\b'  # Matches any 4-digit year from 1900–2099
da = {}

for i, file in enumerate(file_list):
    # Find all years in the filename
    years = re.findall(year_pattern, file)
    
    if years:
        # Extract full 4-digit years using a refined pattern
        full_years = re.findall(r'\b(?:19|20)\d{2}\b', file)
        # Use the first year (e.g., start year of the dataset)
        year = full_years[0]
        print("Extracted year:", year)
        da[f"da{year}"] = xr.open_dataset(file)
    else:
        print(f"Year not found in file: {file}")
```

```python
for key in da.keys():
    if 'DINO' in da[key].data_vars:
        da[key] = da[key].drop_vars(['DINO'])
```

```python
for key in da:
    print(key, da[key].dims, da[key].data_vars)
```

```python
def coarse(file_list):
    # Read files from the file list and assign to a dictionary
    da = {}
    for i, file in enumerate(file_list):
        var_name = f"da{i}" 
        var_value = xr.open_dataset(file)
        da[var_name] = var_value
    
    # Coarsen the data to 1deg X 1 deg grid
    minx = -89.5
    maxx = 89.5
    miny = -179.5
    maxy = 179.5

    new_grid_x = np.arange(minx, maxx+1, 1)
    new_grid_y = np.arange(miny, maxy+1, 1)

    # interpolate using nearest neighbor (can use linear, etc. if desired)
    for d in da:
        # print(d)
        da[d] = da[d].interp(lon=new_grid_y, lat=new_grid_x, method="linear")  
        # print(da[d])
        da[d] = da[d][['CHL','DIATO','MICRO','NANO','PICO']]
    return da
```

```python
# Coarsen the data to 1deg X 1 deg grid
minx = -89.5
maxx = 89.5
miny = -179.5
maxy = 179.5

for key in da.keys():
    # Check the correct latitude and longitude names
    if 'latitude' in da[key].coords and 'longitude' in da[key].coords:
        # Logic to check sizes
        if (da[key].coords['latitude'].size > 180 and da[key].coords['longitude'].size > 360):
            new_grid_y = np.arange(miny, maxy + 1, 1)
            new_grid_x = np.arange(minx, maxx + 1, 1)
            da[key] = da[key].interp(longitude=new_grid_y, latitude=new_grid_x, method="linear")
            da[key] = da[key][['CHL', 'DIATO', 'MICRO', 'NANO', 'PICO']]
            da[key] = da[key].rename({'longitude':'lon','latitude':'lat'})
        else:
            print(f"Skipping dataset {key} due to insufficient lat/lon sizes.")
    else:
        print(f"Latitude or Longitude not found in dataset {key}.")
```
